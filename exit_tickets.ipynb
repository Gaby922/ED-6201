```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager
```

```python
exit_ticket_df = pd.read_csv('data/exit_ticket.csv')
exit_ticket_df
```

```python


# Filter for Mixed-Ability Grouping only
mixed_ability_df = exit_ticket_df[exit_ticket_df['strategy_used'] == "Mixed-Ability Groupings"]

# Ensure all values 1â€“5 are included, even if some are missing
full_range = pd.Series([1, 2, 3, 4, 5])
q1_full_counts = mixed_ability_df['Question 1'].value_counts().reindex(full_range, fill_value=0)
total_responses = q1_full_counts.sum()

# Calculate percentages
percentages = (q1_full_counts / total_responses) * 100

# Define bubble sizes and colors
base_size = 1000
bubble_sizes = percentages * 150 + base_size
bubble_colors = ['#e8344c', '#fd7e1e', '#ffcf2b', '#adda18', '#4ecd56']

# Create the bubble chart
fig, ax = plt.subplots(figsize=(9, 6))

scatter = ax.scatter(
    full_range,
    [1] * 5,
    s=bubble_sizes,
    c=bubble_colors,
    edgecolors="#2e2e2e",
    linewidth=2,
    marker='o'  # Force circular bubbles
)

# Add value and percentage labels
for value, percent in percentages.items():
    ax.text(value, 1.02, str(value), fontsize=16, ha='center', va='center', fontweight='bold', fontname='Andale Mono')
    ax.text(value, 0.98, f"{percent:.1f}%", fontsize=14, ha='center', va='center', fontname='Andale Mono')

# Style the chart
ax.set_xlim(0.5, 5.5)
ax.set_ylim(0.95, 1.05)
ax.set_title("Student Enjoyment Rating of Mixed-Ability Grouping Lesson\n", fontsize=18, fontweight='bold', fontname='Andale Mono')
ax.yaxis.set_visible(False)
ax.xaxis.set_visible(False)
ax.grid(False)
ax.tick_params(left=False, bottom=False)

plt.tight_layout()
plt.savefig('graphs/likert_mixed_ability', dpi=300, bbox_inches='tight')
plt.show()

```

```python
from scipy.stats import kruskal

# Extract groups
group1 = exit_ticket_df[exit_ticket_df["strategy_used"] == "Mixed-Ability Groupings"]["Question 1"]
group2 = exit_ticket_df[exit_ticket_df["strategy_used"] == "Tiered Tasks"]["Question 1"]
group3 = exit_ticket_df[exit_ticket_df["strategy_used"] == "Scaffolding"]["Question 1"]

# Run test
stat, p = kruskal(group1, group2, group3)
print(f"Kruskal-Wallis H Test: H-stat={stat:.3f}, p-value={p:.4f}")

# Interpret
if p < 0.05:
    print("Significant difference between at least two strategies.")
else:
    print("No statistically significant difference found.")
```
